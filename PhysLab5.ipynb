```python
import numpy as np
import matplotlib.pyplot as plt

# create 12 x and y starting points
x = np.linspace(0,11,12)
y = np.linspace(0,11,12)

# map origin of each vector to x,y
(x,y) = np.meshgrid(x,y)

# import x component of vector
Ex = np.loadtxt("Path...ExDipole.txt")

# import y component of vector
Ey = np.loadtxt("Path...EyDipole.txt")

# create and show vector plot
plt.quiver(x,y,Ex,Ey)
plt.show()
```
